```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,200)
y = np.sin(x)

fig, ax = plt.subplots() # the simplest way of creating a fighure with an axes
ax.plot(x, y) # use this to draw some data on the axes
plt.show()
```

```python
fig = plt.figure() # an empty figure with no axes
fig, ax = plt.subplots() # a figure with a simple axes
fig, axs = plt.subplots(2,2) # a figure with 2X2 grid of axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'], ['left','right_bottom']])
```

```python
x = np.linspace(0,2,100)
fig,ax = plt.subplots(figsize = (5,2.7), layout = 'constrained')
ax.plot(x,x,label = 'linear')
ax.plot(x,x**2,label = 'quadratic')
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple Plot")
ax.grid()
ax.legend()
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
# if you need to make the same plots over and over again with different data sets,
# it is convenient to define a function
def my_plotter(ax, data1, data2, param):
    out = ax.plot(data1, data2, **param)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (5,2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data1, data2, {'marker':'o'})
```

```python
fig, ax = plt.subplots(figsize = (5,2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color = 'blue', linewidth = 3, linestyle = '--')
l, = ax.plot(x, np.cumsum(data2),color = 'orange', linewidth = 2)
# l, = 
l.set_linestyle(':')
```

```python
fig, ax= plt.subplots(figsize=(5,2.7))
ax.scatter(data1,data2, s = 50, facecolor = 'C0',edgecolors='k')
```

```python
# marker style
fig, ax = plt.subplots(figsize = (5,2.7))
ax.plot(data1, 'o', label = 'data1')
ax.plot(data2, 'd', label = 'data1')
ax.plot(data3, 'v', label = 'data1')
ax.plot(data4, 's', label = 'data1')
ax.legend()
```

```python
mu, sigma = 115,15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize = (5,2.7), layout = 'constrained')

n, bins, patches = ax.hist(x, 50, density = True, facecolor='C0', alpha = 0.75)

t = ax.set_xlabel('my data', fontsize=14, color='red')
ax.set_ylabel('Probability')
ax.set_title("test")
ax.text(75,.025,r'$\mu = 115,\ \sigma = 15$')
ax.axis([55,175,0,.03])
ax.grid(True)
```
